####  2.2. Why NumPy?

Python lists are powerful, but they werent designed with high-performance computing in mind.

If you try to add two lists together, Python doesnt sum them element-by-elementit concatenates them.

````python
[1, 2, 3 ] + [1, 2, 3]
````

Also, lists dont support advanced numerical operations or linear algebra, and theyre slow for large-scale calculations.

---
**Enter NumPy**

>the Numerical Python library that gives us a powerful object called the **ndarray** (N-dimensional array).
>
> Its like a supercharged listbut:
> * much faster,
> * more memory-efficient, and
> * tailor-made for numerical computations.
---

````python
import numpy as np

# A simple 1D array
arr = np.array([1, 2, 3])
print(arr * 2)
````

This is quite elegant compared to lists! No for loop. Just simple, vectorized math.

####  2.3. Lists vs ndarrays  Whats the difference?

````python
# Python list
my_list = [1, 2, 3]
print([x * 2 for x in my_list])  # You need a loop or comprehension

# NumPy array
my_array = np.array([1, 2, 3])
print(my_array * 2)  # Fast and intuitive
````

Behind the scenes, NumPy arrays use compact memory and C-based operations that are much faster than Python's interpreted loops **[Numpy is written in C, a compiled language, while Python is an interpretted language]**.

In Data Science, where we often process thousands or millions of values, that speed matters.

Additionally, most data we handle comes in 2D form:

####  3.1. What is Pandas?

---

>
> * **Series**: a one-dimensional labeled array (like a column of data)
> * **DataFrame**: a two-dimensional labeled table (like an Excel sheet)
---
Think of pandas as Excelbut in code, and way more powerful.

#### 3.2. From Lists to Series

If a NumPy array is like a list of values, **a Series is like a list with labels (or an index)**.

````python
import pandas as pd

data = [100, 200, 300]

series0 = pd.Series(data) # without specifying indices

series1 = pd.Series(data, index=['Alice', 'Bob', 'Charlie']) # indices are specified

print(series0, "\n")
print(series1)
````

Notice how the data is indexed by default if indices are not provided.

This label-based indexing is one of the things that makes pandas so intuitive.

####  3.3. From ndarrays to DataFrames

A DataFrame is the star of the show in pandas.
Think of it as a full spreadsheet where each column can be of a different type, and each row has an index.

````python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Score': [85.5, 90.2, 78.9]
}

df0 = pd.DataFrame(data)
df1 = pd.DataFrame(data, index=['a.lice', 'b.ob', 'c.harlie'])

print(df0, "\n")
print(df1)
````

Notice how it automatically assigned row indices(`df0`).

Similar to series, you can also assign row indices to dataframes (`df1`)

Each column is like a Series.

You can access any column easily, in two ways:

````python
print(df0['Name'], "\n")  # bracket notation
print(df1.Age)  # dot notation
````

#### 4.1. PySpark  Distributed Data Processing at Scale

PySpark is the Python API for Apache Spark, a big data processing engine that can handle massive datasets across many machines (clusters).

Unlike pandas, which loads data into memory on a single machine, PySpark **distributes the data and computation across many nodes**, allowing you to work with **terabytes** or even **petabytes** of data.

PySpark uses a DataFrame API that looks and feels similar to pandas.

It's heavily used in industry where data is big and speed is critical  think large companies, streaming data, or logs.

##### 4.1.1. PySpark Use Cases

* Processing large logs and event data (like website clicks, server logs)
* Real-time streaming and ETL pipelines
* Machine learning on huge datasets using MLlib (Sparks ML library)
* Data engineering workflows on cloud platforms like AWS, Azure, or Databricks

##### 4.1.2. Basic PySpark Example

````python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Example").getOrCreate()

df = spark.read.csv('large_dataset.csv', header=True, inferSchema=True)

df.show(5)  # Preview data

df.filter(df['age'] > 30).groupBy('city').count().show()
````

#####  4.1.3. Understanding the Example

```
from pyspark.sql import SparkSession
```
This line **imports the `SparkSession` class** from PySpark.
Think of `SparkSession` as the **entry point** for working with Spark. It's how you connect to Spark and start doing data operations.


```
spark = SparkSession.builder.appName("Example").getOrCreate()
```

This line **creates (or retrieves) a Spark session** and names the application `"Example"`.

* `.builder`: sets up the configuration.
* `.appName("Example")`: gives your Spark job a name (useful for logging).
* `.getOrCreate()`: returns an existing Spark session if one exists; otherwise, it creates a new one.

 You now have a **`spark`** object to use for reading and processing data.


```
df = spark.read.csv('large_dataset.csv', header=True, inferSchema=True)
```

This line **reads a CSV file** into a Spark DataFrame (`df`).

* `'large_dataset.csv'`: path to your CSV file.
* `header=True`: tells Spark the **first row contains column names**.
* `inferSchema=True`: Spark will **automatically detect data types** (e.g. int, float, string).

 Result: A Spark DataFrame (like a table), ready to be queried.


```
df.show(5)  # Preview data
```

This line **displays the first 5 rows** of the DataFrame.
Useful for **quickly inspecting** the data structure and content.


```
df.filter(df['age'] > 30).groupBy('city').count().show()
```
This line performs a **mini data pipeline** in Spark:

1. **`filter(df['age'] > 30)`**: keeps only rows where the `age` column is **greater than 30**.
2. **`groupBy('city')`**: groups the filtered data by the `city` column.
3. **`count()`**: counts how many records are in each group (i.e. number of people over 30 in each city).
4. **`show()`**: displays the result.

#### 4.2. Polars  The Lightning-Fast Alternative to Pandas

Polars is a newer DataFrame library designed for **speed and efficiency on a single machine**.

Its written in Rust under the hood, which means its much faster than pandas for many operations.

Its popularly used for **handling moderately large datasets** that pandas struggles with due to speed or memory.

It has a syntax similar to pandas, so switching is straightforward.

##### 4.2.1. Polars Use Cases

* Speeding up your pandas workflows without moving to a distributed system
* Processing datasets too large for pandas to handle efficiently

| -------------- | --------------------------------------------------------------------- | ---------------------------------------------------- |
|  **Pandas**  | Small to medium datasets, quick and flexible analysis                 | Memory-bound, slower on very large datasets          |

####  5.1. Enter GeoPandas: Pandas + Geography

**GeoPandas** lets you do all the things you'd do with pandas**filtering, grouping, joining, and analyzing**but now on **geographic features** like:

*  **Points** (e.g., store locations)
*  **Lines** (e.g., roads, rivers)
*  **Polygons** (e.g., administrative boundaries)

Example spatial dataset:

| Name          | Type    | Geometry                                                              |
|---------------|---------|-----------------------------------------------------------------------|
| Sensor A      | point   | `POINT (1.00000 1.00000)`                                             |
| Main Road     | line    | `LINESTRING (0.00000 0.00000, 2.00000 2.00000, 3.00000 1.00000)`       |
| Central Park  | polygon | `POLYGON ((0.00000 0.00000, 2.00000 0.00000, 2.00000 2.00000, 0.00000 2.00000, 0.00000 0.00000))` |



####  5.2. GeoData File Formats

To work with spatial data, youll often encounter shapefiles and GeoJSON files:

| Format                             | Description                                                                                                            |
| ---------------------------------- | ---------------------------------------------------------------------------------------------------------------------- |
| **Shapefiles (`.shp`)**            | A common format for geographic data used in GIS systems. Usually comes as a **set of files** (.shp, .shx, .dbf, etc.). |
| **GeoJSON (`.geojson` / `.json`)** | A modern, web-friendly format based on JSON. Easy to read and share. Often used for APIs and web maps.                 |


####  5.3. Loading data with GeoPandas


```
import geopandas as gpd
```
